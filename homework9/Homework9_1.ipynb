```python
all_songs=[]
songs=doc.find_all(class_="chart-list-item__first-row chart-list-item__cursor-pointer")
for song in songs:
    row={}
    row['artist']=song.find('div', class_="chart-list-item__artist").get_text(strip=True)
    row['ranks']=song.find('div', class_="chart-list-item__rank").get_text(strip=True)
    row['title']=song.find('span', class_="chart-list-item__title-text").get_text(strip=True)
    all_songs.append(row)
    
all_songs

```

```python
df=pd.DataFrame(all_songs)
df.head()
                    
```
