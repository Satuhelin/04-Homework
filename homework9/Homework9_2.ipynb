```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors'
response = requests.get(url)
doc = BeautifulSoup(response.text)
```

```python
table=doc.find_all('tr')
table
```

```python
whole_table=[]
tables=doc.find_all('tr')
for table in tables:
    row={}
    row['title']=table.find('span', itemprop="name").text
    row['writer']=table.find('a', class_="authorName").string
    score_vote=table.find('span', class_="smallText uitext").find_all('a')
    row['score']=score_vote[0].string
    row['vote']=score_vote[1].string
    row['rating']=table.find('span', class_='minirating').text

    whole_table.append(row)
    
whole_table
```

```python
df=pd.DataFrame(whole_table)
df.head()
```

```python
df.shape
```

```python
df.to_csv('book_by_women.csv', index=False)
```
